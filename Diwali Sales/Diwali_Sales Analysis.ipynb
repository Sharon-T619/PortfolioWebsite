```python
# Import Libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Read in the dataset
df=pd.read_csv('Diwali Sales Data.csv')
```

```python
# Check the shape of the data
df.shape
```

```python
# See the top 10 entries
df.head(10)
```

```python
# Information on data_ number of columns, entries, datatypes
df.info()
```

```python
# Drop unrelated columns
df.drop(['unnamed'], axis=1, inplace=True)
```

```python
# Check the information again
df.info()
```

```python
# Check for NULL values
pd.isnull(df)
```

```python
#Drop NULL values
df.dropna(inplace=True)
```

```python
# Check the data shape
df.shape
```

```python
# Change data type from float to integer
df['Amount'] = df['Amount'].astype('int')
```

```python
# Check changed data type
df['Amount'].dtype
```

```python
# Find data_columns
df.columns
```

```python
# Rename columns
df.rename(columns={'Occupation':'Service'})
```

```python
# Generate descriptive statistics
df[['Cust_name','Age','Occupation','Amount']].describe()
```

```python
# Variable 1: Gender
ax=sns.countplot(x='Gender',data=df)
for bars in ax.containers:
 ax.bar_label(bars)

 ### Interpretation: Most buyers are Female
```

```python
## Variable 2: Occupation
oc=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
bx=sns.barplot(x='Occupation' ,y='Amount',data=oc)
plt.xticks(rotation=90)
# For bars in bx.containers: bx.bar_label(bars)
plt.show

### Interpretation: The top 5 buyers are from the IT sector, Healthcare sector, Aviation, banking and Government respectively.
```

```python
## Variable 3: Age Group
 
# (a) Age Group and Gender Count
ax=sns.countplot(data=df,x='Age Group',hue='Gender')
for bars in ax.containers:
 ax.bar_label(bars)

## Interpretation: Top clients from both genders belong to the same Age group [26-35 years], clients from the 0-17 years Age Group make up the lowest count of clients
```

```python
# (b) Age Group and Amount
cx=df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(data=cx, x='Age Group',y='Amount')

## Interpretation: 26-35 years Age Group are the most profitable group while 0-17 years are the least profitable group
```

```python
## Variable 4: State
# (a) State and Orders
dx=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
dx
```

```python
# Top 10 States based on Orders made by the customers
sns.set(rc={'figure.figsize':(13,5)})
plt.xticks(rotation=40)
sns.barplot(data=dx , x='State',y='Orders')

## Interpretation: From the graph we can see that most of the orders come from Uttar Pradesh, Maharashtra and Karnataka states respectively
```

```python
# (b) State and Amount
# Top 10 States with most spent Amount by customers 
ex=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.barplot(x='State',y='Amount',data=ex)
plt.xticks(rotation=40)
plt.show

## Interpretation: States with the highest number of orders were also responsible for the highest amounts sold _ [Uttar Pradesh, Maharashtra, Karnataka]
```

```python
## Variable 5: Marital Status
# (a) Married Vs Unamarried Count
ex=sns.countplot(data=df, x='Marital_Status')
sns.set(rc={'figure.figsize':(5,3)})
for bars in ex.containers:
 ex.bar_label(bars)

## Interpretation: Most buyers are Unmarried
```

```python
# (b) Marital_Status, Gender and Amount
mar= df.groupby (['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(5,4)})
sns.barplot(data=mar, x = 'Marital_Status', y='Amount',hue='Gender')

## Interpretation: From the graph we can see that most of the buyers are unmarried women
```

```python
### PRODUCT CATEGORY
## Variable 6: Product Category 
#(a) Product Category Count
cx=sns.countplot(data=df,x='Product_Category')
sns.set(rc={'figure.figsize':(15,6)})
for bars in cx.containers:
 cx.bar_label(bars)
plt.xticks(rotation=80)

## Interpretation: Top sold Products are from: Clothing & Apparel, Food, Electronic & Gadgets respectively while the least sold product is Hand & Power Tools
```

```python
# (b) Product Category and Orders
dx=df.groupby(['Product_Category'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False)
sns.barplot(data=dx,x='Product_Category',y='Orders')
plt.xticks(rotation=90)

## Interpretation: Clothing & Apparel, Food and Electronic & Gadgets have the highest number of orders made respectively
```

```python
# (c) Product_Category and Amount
px=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Product_Category',y='Amount',data=px)
plt.xticks(rotation=70)

## Interpretation: Food, Clothing & Apparel, Electronis & Gadgets and Footwear & shoes Product_Categories produce the top revenue for the busniess
```

```python
## Variable 7: Product_ID and Orders
# Top 10 selling products
sells=df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(12,5)})
sns.barplot(data=sells, x='Product_ID', y='Orders')

## Interpretation: The top selling products are: [Product_ID P00265242, P00110942, P00237542]
```

```python
#### CONCLUSION:
# Single women between the age of 26-35 years from Uttarpradesh,Maharashtra and Karnataka working in IT sector ,Healthcare and Aviation are more likely to buy Products from Food, Clothing & Apparel and Electronic & Gadgets Categories
```
